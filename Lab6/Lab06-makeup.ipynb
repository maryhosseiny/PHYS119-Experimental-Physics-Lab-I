## Part A - Intro & Learning Activities

### Notes From Slides
#### Reduced Chi-squared

1) Start with the residuals
2) Normalize residuals using uncertainty
3) Square the normalized residuals and add them together. We square them so that both positive and negative residuals all add up
4) Divide by the number of data points, N so that a set with more points isn't penalized/ We also want a model with more parameters to be penalized. P is the number of parameters in the model - for a straight line with only an adjustable slope, P=1. If a non-zero intercept is allowed, P=2.

#### Interpreting Reduced Chi-squared Value

- If = 1: The models fit the data well, assuming uncertainties have been characterized well
- If >> 1: Not a good fit or the uncertainties have been underestimated
- If << 1: The uncertainties have been overestimated
- Side Note: When using chi-squared, the goal is to adjust your fitting parameters in order to minimize the value for chi-squared, which indicates the best possible fit of your model to the data.
- Side Note #2: The goal is not to make chi-squared = 1, it is to minimize it to find the best possible fit, and then interpret the resulting chi-squared value. 

#### A Good Fit After Minimization
- two-thirds of measurements within 1 deviation of the residuals zero-line
- Almost all (95%) within 2 deviations of the residuals zero line
- No trend in the data (data seems randomly scattered about the residuals zero line) 

#### Experimental Goal

In last week's lab, we were given the residuals, and we used the residuals as the parameters in our model to determine when we reached the best fit. However, the limitation of that method is that it still requires more interpretation. Using chi-squared we can determine how good our model fit is. We will keep repeating this until we reach the minimized chi-squared (the best fit to the data). 

#### Tips
- Keep track of chi-squared value in a separate table to compare the different models' fits with different slopes
- Try to minimize chi-squared value as close to 1 as possible for the best fit
- Once you find the maximum and minimum best fit for slope value. Find the difference between the slope values and divide by 4 to get the uncertainty. Eg. if Max = 2.2 and Min = 2, uncertainty would be (2.2-2)/4=0.2/4=0.05 uncertainty
